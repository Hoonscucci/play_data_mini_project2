# 미니 프로젝트 회고

## 1. 프로젝트 소개
---
- 프로젝트 주제 : 구매 데이터를 통한 구매 예측 및 쿠폰 추천
- 프로젝트 기간 : 2023.06.12 ~ 2023.06.23

## 2. 프로젝트 과정
---
- 주제선정 : 수업을 통해 배운 머신러닝과 딥러닝 내용을 활용하며 데이터 파이프라인
             구성을 목표로 잡고 캐글에서 일본의 온라인 쇼핑몰 'ponpare몰'의  고객정보와 행동 이력을 기반으로 추천 모델을 만드는 것을 프로젝트의 주제로 선정 하였다.
             
- 데이터의 이해도 부족 : 일본어로 구성된 데이터 이다 보니 지역명을 이해하거나 해석에 있어서 조금 지장이 있었다. 또한 일본의 구매 특성이나 이런것에 대한 파악이 조금 어려웠다.

- 코사인 유사도 : 캐글에서 추천 모델을 사용하는데 코사인 유사도를 사용 하였으나 팀원들이 코사인 유사도에 대한 개념이 익숙하지 않았고 주어진 프로젝트 기간이 길지 않고 수업을 통해 배운 개념을 활용해보기 위해 해당 개념을 사용하지 않게 되었다.

- 테이블 통합 : 데이터를 다운 받았을때 총 10개 정도의 파일로 구성되어 있었고 팀원끼리 분담하여 EDA를 만들고 bigquery환경에서 테이블 join을 하여 테이블을 만들었다.

- 데이터 전처리 : 라벨 인코딩으로 숫자형 데이터로 바꿔주고 숫자형 데이터는 standardscaler를 사용하여 전처리를 진행 하였다.
- 과대 적합 : accuracy 확인을 했는데 0.95점으로 상당히 높은 점수가 나왔으나 타겟값의 비율이 95:5이고 혼동행렬을 확인했을때 구매할 확률에 대한 예측을 전혀 하지 못하고 과대 적합이 발생하여 오버 샘플링을 진행 하였습니다.

- 오버 샘플링 : 오버샘플링을 진행하여 train 데이터의 타겟값을 5:5 비율로 맞춰 주었고 accuracy 확인결과 61점으로 점수는 떨어졌지만 혼동 행렬 확인결과 미미하지만 구매할 것에 대한 예측이 발생했기에 유의미한 결과라고 파악했습니다.

## 3. 아쉬운점
---
- EDA 작성 과정에서 유용한 피쳐를 선택하여 그룹화 해보지 못한것에 대한 아쉬움
- LabelEncoder가 아니라 one-hot 인코더를 사용했으면 어땠을까 하는 아쉬움
- 도메인에 대한 지식 부족으로 인해 더 깊이 있는 분석을 하지 못했음에 대한 아쉬움
- 좀 더 최선을 다하지 못한것에 대해 아쉽다.

## 4. 느낀점
---
- 데이터 분석에 대해 아무것도 몰랐지만 그 진행과정에 대해 알게 되는 시간이였다.
- 분석에 있어 단편적인 고민이 아닌 좀더 깊이 있는 고민을 해야함을 느꼈다.
- 팀원들과 거의 매일 to do list 를 작성하고 체크하며 정리를 하였고 정리한 내용은 마지막 최종 정리에 굉장히 큰 도움을 주었다 이를 통해 정리에 대해 다시금 생각하게 되었다.
- 분석에 있어서 잘 알고 관심있는 도메인이 분석 과정에 있어 더 흥미와 능률을 올릴 수 있을 것 같다는 생각이 들었다.
